```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Training set
inputX = np.array([[1.,3.],[1.,2.],[1.,1.5],
                   [1.5,2.],[2.,3.],[2.5,1.5],
                   [2.,1.],[3.,1.],[3.,2.],
                   [3.5,1.],[3.5,3.]])
inputY = [[1.,0.]]*6+ [[0.,1.]]*5
print(inputX)
print(inputY)
```

```python
yc = [0]*5 + [1]*6
print(yc)
plt.scatter(inputX[:,0],inputX[:,1],c=yc, s=50, alpha=0.9)
plt.show()
```

```python
train_features = tf.convert_to_tensor(inputX)
train_labels = tf.convert_to_tensor(inputY)
```

# Single Layer Perceptron with TensorFlow 2.x

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(2),
    tf.keras.layers.Dense(2)
])
```

```python
model.compile(optimizer='SGD',
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
h = model.fit(train_features, train_labels, epochs=2000)
```

```python
acc_set = h.history['loss']
epoch_set = h.epoch
```

```python
# return list of every 10th item in a larger list
plt.plot(epoch_set[0::100],acc_set[0::100], 'o', label='Training phase')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend()
```

```python
testX = np.array([[1.,2.25],[1.25,3.],
                  [2,2.5],[2.25,2.75],
                  [2.5,3.],[2.,0.9],
                  [2.5,1.2],[3.,1.25],
                  [3.,1.5],[3.5,2.],
                  [3.5,2.5]])
testY = [[1.,0.]]*5 + [[0.,1.]]*6

test_features = tf.convert_to_tensor(testX)
test_labels = tf.convert_to_tensor(testY)
```

```python
yc = [0]*5 + [1]*6
print(yc)
plt.scatter(testX[:,0],testX[:,1],c=yc, s=50, alpha=0.9)
plt.show()
```

```python
test_loss, test_acc = model.evaluate(test_features, test_labels, verbose=2)
```

# Multi Layer Perceptron (with One Hidden Layer) with TensorFlow
